```python
# Reload using the correct header row
nba_df = pd.read_excel('/content/nba.xlsx', header=1)

# Show first few rows
nba_df.head()

```

```python
# Check column names and data types
print("Columns:\n", nba_df.columns)

# Check for missing values
print("\nMissing values:\n", nba_df.isnull().sum())

# Show a statistical summary for numeric features
nba_df.describe()
```

```python
nba_df['winning_record'] = (nba_df['winloss'] > 0.5).astype(int)
```

```python
nba_df['games_won'] = nba_df['games won']
```

```python
print(nba_df.columns.tolist())
```

```python
features = [
    'total attendance', 'totalsocialmediafollowers', 'ppg', 'pointsallowed', ' SRS',
    'avgticket', 'fan engagement', 'teamworth', 'earningpretax', 'jerseysale', 'tvgames'
]
```

```python
X = nba_df[features].copy()
```

```python
y_classification = nba_df['winning_record']
y_regression = nba_df['games_won']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

# Split
X_train, X_test, y_train, y_test = train_test_split(X, y_classification, test_size=0.2, random_state=42)

# Model
clf = LogisticRegression(max_iter=1000)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

# Evaluation
print("Classification Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Model
reg = LinearRegression()
reg.fit(X_train, y_train)
y_pred_reg = reg.predict(X_test)

# Evaluation
print("Regression R2 score:", r2_score(y_test, y_pred_reg))
print("Regression MSE:", mean_squared_error(y_test, y_pred_reg))
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# KMeans
kmeans = KMeans(n_clusters=3, random_state=42)
nba_df['cluster'] = kmeans.fit_predict(X_scaled)

# Show clusters
nba_df[['TEAM', 'cluster']].head()
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns

# Reduce to 2 components
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Add PCA components to DataFrame
nba_df['PCA1'] = X_pca[:, 0]
nba_df['PCA2'] = X_pca[:, 1]

# Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=nba_df,
    x='PCA1',
    y='PCA2',
    hue='cluster',
    palette='Set1',
    s=100
)
plt.title('Team Clusters (PCA Projection)')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.legend(title='Cluster')
plt.show()
```
