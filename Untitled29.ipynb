```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

from prophet import Prophet

from nba_api.stats.endpoints import playercareerstats, commonplayerinfo
from nba_api.stats.static import players
```

```python
from nba_api.stats.endpoints import commonplayerinfo
from datetime import datetime

def get_player_stats(player_name):
    # Get player ID
    player_id = get_player_id(player_name)

    # Get career stats
    career = playercareerstats.PlayerCareerStats(player_id=player_id)
    stats_df = career.get_data_frames()[0]
    stats_df = stats_df[stats_df['LEAGUE_ID'] == '00']  # NBA only

    # Get birth year
    info = commonplayerinfo.CommonPlayerInfo(player_id=player_id)
    birth_date_str = info.get_data_frames()[0]['BIRTHDATE'].values[0]
    birth_year = int(birth_date_str.split('-')[0])

    # Extract season start year and calculate age
    stats_df['SEASON_START'] = stats_df['SEASON_ID'].apply(lambda s: int(s[:4]))
    stats_df['AGE'] = stats_df['SEASON_START'] - birth_year

    # Select columns
    stats_df = stats_df[['SEASON_ID', 'SEASON_START', 'AGE', 'GP', 'PTS', 'AST', 'REB', 'FG_PCT', 'FG3_PCT', 'MIN']]
    return stats_df
```

```python
tatum_df = get_player_stats('Jayson Tatum')
edwards_df = get_player_stats('Anthony Edwards')

tatum_df['PLAYER'] = 'Tatum'
edwards_df['PLAYER'] = 'Edwards'

df = pd.concat([tatum_df, edwards_df])
df.reset_index(drop=True, inplace=True)
df.head()
```

```python
features = ['PTS', 'AST', 'REB', 'FG_PCT', 'FG3_PCT', 'MIN']
df_pca = df.dropna(subset=features)  # drop rows with missing values

# Standardize features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df_pca[features])

# Run PCA
pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled_features)

# Add PCA results to DataFrame
df_pca['PCA1'] = pca_result[:, 0]
df_pca['PCA2'] = pca_result[:, 1]
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df_pca, x='PCA1', y='PCA2', hue='PLAYER', style='SEASON_ID', s=100)

for i in range(len(df_pca)):
    plt.text(df_pca['PCA1'].iloc[i]+0.1, df_pca['PCA2'].iloc[i], df_pca['SEASON_ID'].iloc[i], fontsize=8)

plt.title("Career Trajectories in PCA Space")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.grid(True)
plt.show()
```

```python
from prophet import Prophet

# Filter only Edwards' data
edwards_ppg = df[df['PLAYER'] == 'Edwards'][['SEASON_START', 'PTS']].copy()
edwards_ppg.rename(columns={'SEASON_START': 'ds', 'PTS': 'y'}, inplace=True)

# Prophet requires a datetime, so convert season start to actual date
edwards_ppg['ds'] = pd.to_datetime(edwards_ppg['ds'], format='%Y')

# Fit model
model_edwards = Prophet()
model_edwards.fit(edwards_ppg)

# Make future dataframe (e.g. next 3 years)
future_edwards = model_edwards.make_future_dataframe(periods=3, freq='Y')
forecast_edwards = model_edwards.predict(future_edwards)

# Plot
model_edwards.plot(forecast_edwards)
plt.title("Anthony Edwards - Forecasted PPG")
plt.xlabel("Season")
plt.ylabel("PPG")
plt.grid(True)
plt.show()
```

```python
def radar_chart(stats1, stats2, labels, name1, name2):
    import matplotlib.pyplot as plt
    from math import pi

    # Convert to list if needed
    stats1_ = list(stats1[:])
    stats2_ = list(stats2[:])

    categories = list(labels)
    N = len(categories)

    # Angle for each axis
    angles = [n / float(N) * 2 * pi for n in range(N)]
    angles += angles[:1]

    # Close the radar plot
    stats1_.append(stats1_[0])
    stats2_.append(stats2_[0])

    # Create radar chart
    fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))
    plt.subplots_adjust(top=1.1, bottom=0.05)

    # Plot each player
    ax.plot(angles, stats1_, linewidth=2, linestyle='solid', label=name1, color='tab:blue')
    ax.plot(angles, stats2_, linewidth=2, linestyle='dashed', label=name2, color='tab:orange')
    ax.fill(angles, stats1_, color='tab:blue', alpha=0.15)
    ax.fill(angles, stats2_, color='tab:orange', alpha=0.15)

    # Labels on axes
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(categories, fontsize=12)

    # Circle axis labels
    ax.set_rlabel_position(30)
    ax.set_yticks([0.2, 0.4, 0.6, 0.8])
    ax.set_yticklabels(["0.2", "0.4", "0.6", "0.8"], fontsize=10, color='gray')
    ax.set_ylim(0, 1.0)

    # Add legend
    ax.legend(loc='upper right', bbox_to_anchor=(1.1, 1.1), fontsize=12)
    ax.set_title(f"{name1} vs {name2} — First 7 Seasons", fontsize=14, pad=30)

    plt.show()
```

```python
from nba_api.stats.endpoints import playercareerstats
from nba_api.stats.static import players
import pandas as pd

# Find player ID by name
def get_player_id(name):
    return players.find_players_by_full_name(name)[0]['id']

# Pull basic season stats for a player
def get_basic_stats(player_name):
    player_id = get_player_id(player_name)
    career = playercareerstats.PlayerCareerStats(player_id=player_id)
    df = career.get_data_frames()[0]
    df = df[df['LEAGUE_ID'] == '00']  # NBA only
    df = df[['SEASON_ID', 'TEAM_ID', 'GP', 'PTS', 'AST', 'REB', 'FG_PCT', 'FG3_PCT']]
    df['PLAYER'] = player_name
    return df.reset_index(drop=True)
```
